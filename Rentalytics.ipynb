```python
import pandas as pd
import numpy as np
```

```python
#Load the data into a pandas dataframe
df = pd.read_csv("data/TorontoListings.csv")
df
```

```python
#Drop columns not needed for analysis
df = df.drop(["id", "listing_url", "scrape_id", "last_scraped", "source", "picture_url", "host_url", "host_name", "host_thumbnail_url", "host_picture_url", "host_verifications", "host_has_profile_pic", "neighbourhood", "neighbourhood_group_cleansed", "bathrooms", "minimum_minimum_nights", "minimum_maximum_nights", "maximum_minimum_nights", "maximum_maximum_nights", "minimum_nights_avg_ntm", "maximum_nights_avg_ntm", "calendar_updated", "calendar_last_scraped", "number_of_reviews_ltm", "number_of_reviews_l30d"], axis=1)
df
```

```python
# Convert the date column to datetime objects
df["host_since"] = pd.to_datetime(df["host_since"])
```

```python
# Remove listings with missing values in important columns
important_cols = ["neighbourhood_cleansed", "property_type", "room_type", "accommodates", "bedrooms", "beds", "price", "latitude", "longitude"]
df = df.dropna(subset = important_cols)
```

```python
# Use one-hot encoding to convert categorical columns into numerical values
categorical_cols = ["neighbourhood_cleansed", "property_type", "room_type"]
df = pd.get_dummies(df, columns = categorical_cols)
```

```python
# Normalize numerical columns
numerical_cols = ["latitude", "longitude", "price", "minimum_nights", "maximum_nights", "calculated_host_listings_count", "accommodates", "bedrooms", "beds", "number_of_reviews", "review_scores_rating", "review_scores_accuracy", "review_scores_cleanliness", "review_scores_checkin", "review_scores_communication", "review_scores_location", "reviews_per_month"]
df[numerical_cols] = (df[numerical_cols] - df[numerical_cols].mean()) / df[numerical_cols].std()
```
